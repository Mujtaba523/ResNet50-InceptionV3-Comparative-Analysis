## **INTRODUCTION**
* ## What is CNN?

<p align="justify">
A convolutional neural network (CNN) is a type of artificial neural network specifically designed for image classification tasks. CNNs are trained using labeled images and learn to recognize patterns and features in images through a process called convolution. CNNs consist of a series of layers, including convolutional layers, pooling layers, and fully connected layers. The convolutional layers apply a set of filters to the input image, producing a set of filtered images called feature maps. The pooling layers reduce the spatial dimensions of the feature maps, while the fully connected layers combine the features extracted by the convolutional and pooling layers to make a final prediction. </p>

## **CNN ALGORITHMS SELECTION**
* ## ResNet50

<p align="justify">
ResNet (short for "Residual Network") is a convolutional neural network (CNN) architecture developed by He et al. in 2015. ResNet is known for introducing the concept of "shortcut connections," which allow the network to learn residual functions and make it easier to train very deep networks. </p>

<p align="justify">
All of the earlier models made use of deep neural networks, in which numerous convolution layers were stacked one on top of the other. Deeper networks perform better, as was discovered. But it turns out that this isn't actually the case. The issues with deeper networks are as follows:</p>

1. Network optimization becomes challenging
2. Disappearing or exploding gradients
3. Degradation Issue: First, accuracy saturates, then it degrades

<p align="justify">
In order to solve these issues, the creators of the resnet design proposed the concept of skip connections under the presumption that deeper layers should be able to learn just as much as shallower layers. Copying the activations from shallower layers and setting other layers to identity mapping are potential solutions. Skip connections, which are seen in the following image, are what make these linkages possible.</p>

<figure align="center">
  <img src="https://cdn-images-1.medium.com/max/987/1*pUyst_ciesOz_LUg0HocYg.png">
  <figcaption><em>Image 2: Skip Connections</em></figcaption>
</figure>

<p align="justify">
One of the main advantages of ResNet is that it can effectively handle the vanishing gradients problem, which occurs when the gradients of the parameters in the network become very small and the network is unable to learn effectively. This is achieved by the use of shortcut connections, which allow the network to directly propagate the input through the network, bypassing multiple layers. This allows the network to learn more complex features and makes it easier to train very deep networks without suffering from diminishing performance.  </p>

These are the main reasons behind selecting ResNet for comparitive analysis.

* ## VGG16

<p align="justify">
VGG16 is a convolutional neural network (CNN) model trained on the ImageNet dataset. It was developed by the Visual Geometry Group (VGG) at the University of Oxford and introduced in the paper "Very Deep Convolutional Networks for Large-Scale Image Recognition" by Karen Simonyan and Andrew Zisserman in 2014.

VGG16 is a deep CNN model with 16 weight layers, consisting of 13 convolutional layers and 3 fully-connected layers. It has a simple architecture with a small number of filters per layer, and it uses max pooling to reduce the dimensionality of the input data.
</p>



<p align="justify">
VGG16 has been widely used as a base model for transfer learning, where the model is pre-trained on a large dataset and then fine-tuned on a smaller dataset for a specific task. It has also been used as a benchmark model to compare the performance of other CNN models on various image classification tasks.
One reason to use VGG16 is that it has a relatively simple architecture and is easy to implement, which makes it a good choice for beginners. It is also widely available in many deep learning frameworks, including Keras, TensorFlow, and PyTorch.</p>

## **TRAINING/TESTING OF ALGORITHMS**

<p align="justify">
Following process is done in training/testing phase of both algorithms.</p>
<p align="justify">
1. Importing libraries, Defining function which will calculate the execution/training time of algorithms. Hyperparameters are set. Dataset is loaded and then it is split into training set, validation set and testing set. Some images are plotted from the training set.</p>

<p align="justify">
2. Data augmentation is done. Data augmentation is a technique used to artificially increase the size of a dataset by generating additional data samples from the existing data. This can be useful when the dataset is small or not diverse enough to train a robust machine learning model. Data augmentation can be achieved by applying a set of transformations to the existing data, such as cropping, rotating, flipping, scaling, and shifting. These transformations are applied randomly to the data samples to create new, augmented samples that are similar but not identical to the original data.</p>

<p align="justify">
3. ResNet50 Model and VGG16 Model are used through transfer learning. Transfer learning is a machine learning technique where a model trained on a large dataset is re-purposed for a different task by fine-tuning it on a smaller dataset. Transfer learning can be a useful approach when the dataset for the new task is small, or when the new task is similar to the original task for which the model was trained. In transfer learning, a pre-trained model is used as a starting point and the final layers are modified or added to fit the new task. The pre-trained model is called the base model, and the added layers are called the top layers. The base model is typically trained on a large dataset such as ImageNet, which contains millions of images and hundreds of thousands of classes. The base model is then fine-tuned on a smaller dataset for the new task. </p>

```python
# Importing Libraries
from keras.applications import ResNet50, VGG16
from keras.datasets import cifar100
from keras.layers import Dense, Dropout, GlobalAveragePooling2D, BatchNormalization, Flatten 
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.models import Model
from keras.optimizers import Adam
from keras.utils import np_utils
import cv2
import time
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
# Function for execution time calculation
def execution_time(start,end):
  return end-start
```

```python
# Hyperparameters
BATCH_SIZE = 32
LEARNING_RATE = 0.0001
EPOCHS = 50
NUMBER_OF_CLASSES = 100
```

```python
# Loading data
(X_train, y_train), (X_test, y_test) = cifar100.load_data()
# Train-Test Split
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, stratify=np.array(y_train), random_state=42)
# Encoding classes
Y_train = np_utils.to_categorical(y_train, NUMBER_OF_CLASSES)
Y_val = np_utils.to_categorical(y_val, NUMBER_OF_CLASSES)
Y_test = np_utils.to_categorical(y_test, NUMBER_OF_CLASSES)
```

```python
# Plotting 50 images from training data
plt.figure(figsize=(12,12))
for i in range(50):
  plt.subplot(10,10,1+i)
  plt.axis('off')
  plt.imshow(X_train[i])
```

```python
# Data augmentation
train_datagen = ImageDataGenerator(rescale=1./255, horizontal_flip=False)
train_datagen.fit(X_train)
train_generator = train_datagen.flow(X_train,Y_train,batch_size=BATCH_SIZE)
val_datagen = ImageDataGenerator(rescale=1./255,horizontal_flip=False)
val_datagen.fit(X_val)
val_generator = val_datagen.flow(X_val,Y_val,batch_size=BATCH_SIZE)
```

## ResNet50

```python
# ResNet50 Model Building
resnet = ResNet50(weights="imagenet", include_top=False, input_shape=(32, 32, 3))
# Freeze the layers of the model
for layer in resnet.layers:
     layer.trainable = False
# Adding classification layer 
model1 = Sequential()
model1.add(resnet)
model1.add(Flatten())
model1.add(Dense(256,activation="relu"))
model1.add(Dense(256,activation="relu"))
model1.add(Dropout(0.6))
model1.add(Dense(NUMBER_OF_CLASSES,activation="softmax"))
```

```python
# Compiling the model
model1.compile(loss='categorical_crossentropy',optimizer=Adam(learning_rate=LEARNING_RATE),metrics=['accuracy'])
```

```python
# Checking Model Summary
model1.summary()
```

```python
# Training the model
start = time.time()
training_steps_per_epoch = X_train.shape[0] // BATCH_SIZE
validating_steps_per_epoch = X_val.shape[0] // BATCH_SIZE
history1 = model1.fit(train_generator,steps_per_epoch=training_steps_per_epoch, validation_data=val_generator,validation_steps=validating_steps_per_epoch, epochs=EPOCHS, verbose=1)
end = time.time()
execution_time_resnet = execution_time(start,end) 
```

```python
# Testing the model
test_loss, test_acc = model1.evaluate(X_test, Y_test)
print('Test accuracy:', test_acc*100)
```

## VGG16

```python
# VGG16 Model Building
model = VGG16(weights="imagenet", include_top=False, input_shape=(32, 32, 3))
# Freeze the layers of the model
for layer in model.layers:
     layer.trainable = False
# Adding classification layer 
x = GlobalAveragePooling2D()(model.output)
x= BatchNormalization()(x)
x = Dense(256, activation='relu')(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.6)(x)
output = Dense(NUMBER_OF_CLASSES, activation='softmax')(x)
model2 = Model(model.input, output)
```

```python
# Compiling the model
model2.compile(loss='categorical_crossentropy',optimizer=Adam(learning_rate=LEARNING_RATE),metrics=['accuracy'])
```

```python
# Checking Model summary
model2.summary()
```

```python
# Training the Model
start = time.time()
training_steps_per_epoch = X_train.shape[0] // BATCH_SIZE
validating_steps_per_epoch = X_val.shape[0] // BATCH_SIZE
history2 = model2.fit(train_generator,steps_per_epoch=training_steps_per_epoch, validation_data=val_generator,validation_steps=validating_steps_per_epoch, epochs=EPOCHS, verbose=1)
end = time.time()
execution_time_vgg = execution_time(start,end) 
```

```python
# Testing the model
test_loss, test_acc = model2.evaluate(X_test, Y_test)
print('Test accuracy:', test_acc*100)
```

Training time is the amount of time it takes to train a machine learning model using a given dataset. Training time is typically measured in seconds, minutes, or hours, and is used to evaluate the efficiency and performance of a model and its training process.</p>

```python
print("Execution/Training time of ResNet50: ", execution_time_resnet)
print("Execution/Training time of ResNet50: ", execution_time_vgg)
```

* ### *Accuracy and Loss* 
<p align="justify">
Accuracy is a metric used to evaluate the performance of a machine learning model on a given dataset. It is defined as the proportion of correct predictions made by the model out of all the predictions made. Loss is a metric that measures the error or difference between the predicted values and the true values</p>

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
t = fig.suptitle('ResNet50 Performance', fontsize=12)
fig.subplots_adjust(top=0.85, wspace=0.3)

epochs = list(range(1,EPOCHS+1))
ax1.plot(epochs, history1.history['accuracy'], label='Train Accuracy')
ax1.plot(epochs, history1.history['val_accuracy'], label='Validation Accuracy')
ax1.set_xticks(epochs)
ax1.set_ylabel('Accuracy Value')
ax1.set_xlabel('Epoch')
ax1.set_title('Accuracy')
l1 = ax1.legend(loc="best")

ax2.plot(epochs, history1.history['loss'], label='Train Loss')
ax2.plot(epochs, history1.history['val_loss'], label='Validation Loss')
ax2.set_xticks(epochs)
ax2.set_ylabel('Loss Value')
ax2.set_xlabel('Epoch')
ax2.set_title('Loss')
l2 = ax2.legend(loc="best")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
t = fig.suptitle('VGG16 Performance', fontsize=12)
fig.subplots_adjust(top=0.85, wspace=0.3)

epochs = list(range(1,EPOCHS+1))
ax1.plot(epochs, history2.history['accuracy'], label='Train Accuracy')
ax1.plot(epochs, history2.history['val_accuracy'], label='Validation Accuracy')
ax1.set_xticks(epochs)
ax1.set_ylabel('Accuracy Value')
ax1.set_xlabel('Epoch')
ax1.set_title('Accuracy')
l1 = ax1.legend(loc="best")

ax2.plot(epochs, history2.history['loss'], label='Train Loss')
ax2.plot(epochs, history2.history['val_loss'], label='Validation Loss')
ax2.set_xticks(epochs)
ax2.set_ylabel('Loss Value')
ax2.set_xlabel('Epoch')
ax2.set_title('Loss')
l2 = ax2.legend(loc="best")
```
